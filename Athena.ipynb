```python
from pyathena import connect
import pandas as pd
from time import sleep
from collections import Counter
from textblob import TextBlob
import json
import hashlib
import os
import matplotlib.pyplot as plt

# Connect to Athena
conn = connect(
    aws_access_key_id='your access key',
    aws_secret_access_key='your secret key',
    s3_staging_dir='s3://{name of staging directory}/',
    region_name='your region name',
)

def fetch_data():
    """
    Fetch the latest data from Athena.
    """
    query = 'SELECT * FROM "kafka-project-db"."kafka_project_mounes" limit 100;'
    df = pd.read_sql(query, conn)
    return df

def analyze_data(df):
    """
    Perform analysis on the data and return results.
    """
    # 1. Topic Frequency
    topic_counts = df['topic_name'].value_counts().to_dict()

    # 2. Text Length Analysis
    df['text_length'] = df['text'].apply(len)
    text_length_stats = df.groupby('topic_name')['text_length'].agg(['mean', 'min', 'max']).to_dict()

    # 3. Keyword Analysis
    all_words = ' '.join(df['text']).split()
    most_common_words = Counter(all_words).most_common(10)

    # 4. Sentiment Analysis
    df['sentiment'] = df['text'].apply(lambda x: TextBlob(x).sentiment.polarity)
    sentiment_stats = df.groupby('topic_name')['sentiment'].mean().to_dict()

    # Combine all analysis into a dictionary
    analysis = {
        "topic_frequency": topic_counts,
        "text_length_stats": text_length_stats,
        "most_common_words": most_common_words,
        "sentiment_stats": sentiment_stats,
    }

    return analysis

def save_analysis_to_file(analysis, output_dir="analysis_files"):
    """
    Save the analysis to a file and check for changes.
    """
    # Create the output directory if it doesn't exist
    if not os.path.exists(output_dir):
        os.makedirs(output_dir)

    # Generate a hash of the current analysis
    analysis_hash = hashlib.md5(json.dumps(analysis, sort_keys=True).encode()).hexdigest()
    hash_file_path = os.path.join(output_dir, "latest_hash.txt")

    # Check if the hash matches the last saved analysis
    if os.path.exists(hash_file_path):
        with open(hash_file_path, "r") as hash_file:
            last_hash = hash_file.read().strip()
        if last_hash == analysis_hash:
            print("No changes detected in the analysis. Skipping save.")
            return None

    # Save the new analysis
    output_file = os.path.join(output_dir, f"analysis_{analysis_hash[:8]}.json")
    with open(output_file, "w") as file:
        json.dump(analysis, file, indent=4)
    print(f"New analysis saved to {output_file}")

    # Update the latest hash
    with open(hash_file_path, "w") as hash_file:
        hash_file.write(analysis_hash)

    return output_file

def load_latest_analysis(output_dir="analysis_files"):
    """
    Load the newest analysis file from the output directory.
    """
    files = [f for f in os.listdir(output_dir) if f.startswith("analysis_") and f.endswith(".json")]
    if not files:
        print("No analysis files found.")
        return None

    # Get the latest file by sorting by modified time
    latest_file = max(files, key=lambda f: os.path.getmtime(os.path.join(output_dir, f)))
    file_path = os.path.join(output_dir, latest_file)
    with open(file_path, "r") as file:
        analysis = json.load(file)
    print(f"Loaded analysis from {file_path}")
    return analysis

def visualize_analysis_with_ids(analysis):
    """
    Visualize the analysis data with topic IDs for better alignment.
    """
    if not analysis:
        print("No analysis data to visualize.")
        return

    # Assign IDs to topics
    topic_names = list(analysis['topic_frequency'].keys())
    topic_ids = [chr(65 + i) for i in range(len(topic_names))]  # Generate IDs (A, B, C, ...)
    topic_id_mapping = dict(zip(topic_ids, topic_names))

    # Print topic IDs and names for reference
    print("\n--- Topic ID Mapping ---")
    for topic_id, topic_name in topic_id_mapping.items():
        print(f"{topic_id}: {topic_name}")

    # Use topic IDs for plotting
    topic_counts = list(analysis['topic_frequency'].values())
    plt.figure(figsize=(10, 6))
    plt.barh(topic_ids, topic_counts, color='skyblue')
    plt.title("Topic Frequency", fontsize=16)
    plt.xlabel("Number of Entries", fontsize=12)
    plt.ylabel("Topic IDs", fontsize=12)
    plt.tight_layout()
    plt.show()

    # Text Length Statistics
    text_length_stats = analysis['text_length_stats']
    mean_lengths = text_length_stats['mean']
    min_lengths = text_length_stats['min']
    max_lengths = text_length_stats['max']

    x = range(len(mean_lengths))  # X-axis positions
    plt.figure(figsize=(12, 6))
    plt.bar(x, mean_lengths.values(), width=0.3, label="Mean", color='blue', alpha=0.7)
    plt.bar([i + 0.3 for i in x], min_lengths.values(), width=0.3, label="Min", color='green', alpha=0.7)
    plt.bar([i + 0.6 for i in x], max_lengths.values(), width=0.3, label="Max", color='red', alpha=0.7)
    plt.xticks([i + 0.3 for i in x], topic_ids, fontsize=10)  # Use topic IDs
    plt.title("Text Length Statistics", fontsize=16)
    plt.xlabel("Topic IDs", fontsize=12)
    plt.ylabel("Length", fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Sentiment Analysis
    sentiment_stats = analysis['sentiment_stats']
    sentiments = list(sentiment_stats.values())

    plt.figure(figsize=(12, 6))
    plt.bar(topic_ids, sentiments, color='purple', alpha=0.8)
    plt.axhline(0, color='black', linestyle='--', linewidth=1, label="Neutral Sentiment")
    plt.title("Sentiment Analysis (Average Polarity)", fontsize=16)
    plt.xlabel("Topic IDs", fontsize=12)
    plt.ylabel("Sentiment Polarity", fontsize=12)
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Most Common Words
    most_common_words = analysis['most_common_words']
    words = [word[0] for word in most_common_words]
    counts = [word[1] for word in most_common_words]

    plt.figure(figsize=(10, 6))
    plt.bar(words, counts, color='orange', alpha=0.8)
    plt.title("Most Common Words", fontsize=16)
    plt.xlabel("Words", fontsize=12)
    plt.ylabel("Frequency", fontsize=12)
    plt.tight_layout()
    plt.show()



def main():
    """
    Main loop for real-time analysis with change detection and visualization.
    """
    while True:
        print("\nFetching data from Athena...")
        data = fetch_data()

        if not data.empty:
            print("Data fetched. Analyzing...")
            analysis = analyze_data(data)
            new_file = save_analysis_to_file(analysis)

            if new_file:  # If a new file was saved, visualize the data
                analysis = load_latest_analysis()
                visualize_analysis_with_ids(analysis)
        else:
            print("No new data found.")

        # Wait for 1 minute before the next fetch
        print("\nWaiting for the next fetch...")
        sleep(60)

if __name__ == "__main__":
    main()
```
